```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dwelling = pd.read_csv('raw_Dwelling_Units_by_Municipality.csv')
structure = pd.read_csv('raw_housing_buildings_by_municipality.csv')
dwelling.head()
```

```python
dwelling = dwelling.drop(columns=['CSDUID', 'IndicatorSummaryDescription', 'UnitOfMeasure'])
dwelling.rename(columns={'OriginalValue':'Total_Dwelling_Count'}, inplace=True)
dwelling.info()
```

```python
dwelling.to_csv('cleaned_Dwelling_Units_by_Municipality.csv', index=False)
```

```python
structure.head()
```

```python
structure = structure.drop(columns=['CSDUID', 'IndicatorSummaryDescription', 'UnitOfMeasure'])
```

```python
structure = structure.pivot(index=['CSD', 'Period'], columns=['Structural Type'], values='OriginalValue')
structure.head()
```

```python
column_map = {'Apartment, building that has fewer than five storeys':'Apartment < 5 Stories',
              'Apartment, building that has five or more storeys':'Apartment >= 5 Stories',
              'Apartment, duplex':'Duplexes',
              'Movable dwelling':'Movable dwellings',
              'Other attached dwelling':'Attached dwellings',
              'Row house':'Row Houses',
              'Semi-detached house':'Semi-detached houses',
              'Single-detached house':'Single-detached houses'}
```

```python
structure.rename(columns=column_map, inplace=True)
structure.info()
```

```python
structure.to_csv('cleaned_housing_buildings_by_municipality.csv', index=False)
```
